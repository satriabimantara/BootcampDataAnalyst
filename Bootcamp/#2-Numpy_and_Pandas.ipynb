# Bootcamp Numpy - Numpy Array (Indexing and Slicing)

```python
data = [
    ['Denis', 'Scratch', 8],
    ['Frank', 'Python', 18],
    ['Roger', 'Unity', 15],
]


# soal nomor 1
data = np.array(data)

# soal nomor 2
print(data[0])
print(data[1])
print(data[2])

# soal nomor 3
print(data[:,0])
print(data[:,1])
print(data[:,2])

# soal nomor 4
data[0,1] = 'Roblox'
```

```python
data_2 = [
    ['Java', 'Richie', 60, 'C'],
    ['Python', 'Niels', 80, 'A'],
    ['R', 'Brie', 75, 'B'],
]

#soal nomor 1
data_2 = np.array(data_2)

# soal nomor 2
print(data_2[0])
print(data_2[1])
print(data_2[2])

# soal nomor 3
print(data_2[:,0])
print(data_2[:,1])
print(data_2[:,2])
print(data_2[:,2])

# soal nomor 4
data_2[:,0] = "Data Analyst"
print(data_2)
```

# Bootcamp Pandas - Conditional Selection and Modifying DataFrame

```python
data = [
    ['Michael Phelps', 23, 'United States', 2008, 8],
    ['Michael Phelps', 19, 'Canada', 2004, 6],
    ['Michael Phelps', 27, 'United States', 2012, 4],
    ['Natalie Coughlin', 25, 'Norway', 2008, 1],
    ['Aleksey Nemov', 24, 'China', 2000, 2],
]
swimm_df = pd.DataFrame(data, columns=['Athlete', 'Age','Country', 'Year', 'Gold Medals'])
swimm_df
```

```python
# sal nomor 1 tampilkan nama atlet yang berasa dari US
swimm_df[
    swimm_df['Country']=="United States" 
]['Athlete']
```

```python
# soal nomor 2 tampilkan nama atlet yang bergabung pada tahun 2008
swimm_df[
    swimm_df['Year']==2008
]
```

```python
# soal nomor 3 hapus data dengan atlet yang berusia lebih dari 25 tahun dan berasal dari norway
# ambil data yang usia atlet > 25 tahun
filtering_1 = swimm_df[
    swimm_df['Age']>25
]
filtering_2 = filtering_1[
    filtering_1['Country']=="Norway"
]
# ambil index data yang akan dihapus
index = filtering_2.index

# hapus data
new_swimm_df = swimm_df.drop(index)
```

# Bootcamp Data Analysis - Statistika Dasar

```python
data_pelanggaran_lalu_lintas = pd.read_csv('dataset/Data-Penindakan-Pelanggaran-Lalu-Lintas.csv')
data_pelanggaran_lalu_lintas
```

```python
data_pelanggaran_lalu_lintas.dtypes
```

```python
# soal nomor 3 tentukan mean dan median dari setiap masing-masing variabel numerik
print(data_pelanggaran_lalu_lintas['stop_operasi'].mean())
print(data_pelanggaran_lalu_lintas['stop_operasi'].median())
```

```python
# soal nomor 4 tentukan std serta quartile 1,2,3 pada setiap masing-masing variabel numerik
print(data_pelanggaran_lalu_lintas['stop_operasi'].std())
print(np.quantile(data_pelanggaran_lalu_lintas['stop_operasi'], [.25, .5, .75]))
```

```python
# soal nomor 5 tentukan mean dari variabel bap_tilang pada setiap masing-masing bulan
data_pelanggaran_lalu_lintas.groupby(by=['bulan'])[['bap_tilang', 'stop_operasi']].sum()
```

## Violin Plot
***

***
1. In general, violin plots are a method of plotting numeric data and can be considered a combination of the **box plot** with a **kernel density plot**. 
Go to *kernel density estimation (KDE)* : <br> 
- https://en.wikipedia.org/wiki/Kernel_density_estimation
- https://www.itm-conferences.org/articles/itmconf/pdf/2018/08/itmconf_sam2018_00037.pdf
- The Kernel Density Estimation is a mathematic process of finding an estimate probability density function of a random variable.
2. In the violin plot, we can find the same information as in the box plots: 

- median (a white dot on the violin plot)
- interquartile range (the black bar in the center of violin)
- the lower/upper adjacent values (the black lines stretched from the bar) — defined as first quartile — 1.5 IQR and third quartile + 1.5 IQR respectively.

3. The unquestionable advantage of the violin plot over the box plot is that aside from showing the abovementioned statistics it also **shows the entire distribution of the data**.

```python
N = 10 ** 4
np.random.seed(42)
```

```python
def plot_comparison(x, title):
    # create subplot 
    fig, ax = plt.subplots(3, 1, sharex=True)
    sns.histplot(x, ax=ax[0])
    
    # create histogram and kde with x data
    ax[0].set_title('Histogram + KDE')
    sns.boxplot(x, ax=ax[1])
    
    # create boxplot
    ax[1].set_title('Boxplot')
    sns.violinplot(x, ax=ax[2])
    
    # create violin plot
    ax[2].set_title('Violin plot')
    
    # create supertitle and show plot
    fig.suptitle(title, fontsize=16)
    plt.show()
```

### Start With Standard Normal Distribution
***
We draw 10000 numbers at random and plot the results

```python
sample_gaussian = np.random.normal(size=N)
plot_comparison(sample_gaussian, 'Standard Normal Distribution')
```

**Interpretation** <br>
***
1. in the histogram we see the symmetric shape of the distribution
2. we can see the previously mentioned metrics (median, IQR, Tukey’s fences) in both the box plot as well as the violin plot
3. Wider sections of the violin plot represent a higher probability of observations taking a given value, the thinner sections correspond to a lower probability.

### Log Normal Distribution

```python
sample_lognormal = np.random.lognormal(size=N)
plot_comparison(sample_lognormal, 'Log-normal Distribution')
```

### Mixture of Gaussian
***
In the previous two examples, we have already seen that the violin plots contain more information than the box plot. This is even more apparent when we consider a multimodal distribution.

```python
sample_bimodal = np.concatenate([np.random.normal(loc=-2, scale=2, size=int(N/2)),
                                 np.random.normal(loc=3, scale=1, size=int(N/2))])
plot_comparison(sample_bimodal, 'Mixture of Gaussians - bimodal')
```

## Scatter Plot
***
1. Berguna untuk memeriksa korelasi antar variabel secara visual

***
A scatter plot (2D) is a useful method of comparing variables against each other. Scatter plots look similar to line plots in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

***

## Bar Plot
***
- Membandingkan beberapa kuantitas dari beberapa kategori atau kelompok tertentu

***
**Example** <br>
1. Dengan data **PopulationEstimates** lihat data tingkat kelahiran dan tingkat kematian untuk Anderson County di Texas dari tahun 2011 hingga 2018

```python
# get the Anderson County data
anderson_county_df = population_estimate[
    (population_estimate['Area_Name'] == 'Anderson County') &
    (population_estimate['State']=='TX')
].reset_index(drop=True)
anderson_county_df
```

```python
# get the birth and death rates
def check_birth_columns(x):
    return x.startswith('R_birth')
def check_death_columns(x):
    return x.startswith('R_death')

birthRates = list(filter(check_birth_columns, anderson_county_df.columns))
deathRates = list(filter(check_death_columns, anderson_county_df.columns))
years = np.array(list(
    map(
        lambda x: int(x[-4:]),
        birthRates
    )
)
)
```

```python
plt.figure(figsize=(10,6))
width = 0.4
plt.plot(anderson_county_df[birthRates].transpose().values, label='Birth Rate Anderson County Texas', linestyle='-.')
plt.plot(anderson_county_df[deathRates].transpose().values, label='Death Rate Anderson County Texas')
plt.xticks(ticks=range(0, len(years)), labels=years)
plt.legend()
plt.title('Anderson County Birth vs Death Rate in Texas from 2011 - 2018')
plt.show()
```

***
